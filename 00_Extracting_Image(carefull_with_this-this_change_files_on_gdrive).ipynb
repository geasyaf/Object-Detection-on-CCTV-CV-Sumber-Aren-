# Extract to image video by video

jangan lupa setiap fideo dibuatkan folder berbeda agar dapat dinavigasi dengan mudah

```python
import cv2
import os

# Tentukan path file video dan direktori output untuk frame
video_path = file_path + video_ke + videofromat
output_dir = '/content/' + video_ke + '/'

# Buat direktori output jika belum ada
os.makedirs(output_dir, exist_ok=True)

# Fungsi untuk mengekstrak frame dari video
def extract_frames(video_path, output_folder):
    video_name = os.path.basename(video_path).split('.')[0]
    cap = cv2.VideoCapture(video_path)

    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(fps)  # Ambil frame per detik
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % frame_interval == 0:
            frame_file = os.path.join(output_folder, f"{video_name}_frame_{frame_count // frame_interval}.jpg")
            cv2.imwrite(frame_file, frame)

        frame_count += 1

    cap.release()

# Ekstrak frame dari video
extract_frames(video_path, output_dir)

print("Ekstraksi frame selesai!")
```

# Ubah Folder Extracted Image menjadi zip sehingga dapat di download ke komputer masing masing

```python
import shutil
import os

# Tentukan path folder yang ingin di-zip dan nama file zip hasilnya
folder_to_zip = '/content/' + video_ke + '/'
output_zip_file = '/content/' + video_ke + '.zip'

# Buat zip file dari folder
shutil.make_archive(output_zip_file.replace('.zip', ''), 'zip', folder_to_zip)

print(f"Folder {folder_to_zip} berhasil di-zip menjadi {output_zip_file}")
```

# Salin File ZIP ke Google Drive

```python
import shutil
import os

# Tentukan path tujuan di Google Drive
drive_path = '/content/drive/MyDrive/videotoimage/' + video_ke + '.zip'

# Salin file zip ke Google Drive
shutil.copy(output_zip_file, drive_path)
print(f"File {output_zip_file} berhasil diunggah ke {drive_path}")
```
